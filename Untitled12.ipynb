```python

import pandas as pd
import matplotlib.pyplot as plt

# Sample manufacturing sector data
data = {
    'Date': ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01'],
    'Frontend': [85, 88, 90, 86, 89],
    'Backend': [82, 84, 86, 83, 85]
}

# Convert data to DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Sort DataFrame by Date
df = df.sort_values(by='Date')

# Calculate the run chart line
df['Mean_Frontend'] = df['Frontend'].mean()
df['Mean_Backend'] = df['Backend'].mean()

# Plot the run chart
plt.plot(df['Date'], df['Frontend'], marker='o', label='Frontend')
plt.plot(df['Date'], df['Backend'], marker='o', label='Backend')
plt.plot(df['Date'], df['Mean_Frontend'], linestyle='--', color='green', label='Mean Frontend')
plt.plot(df['Date'], df['Mean_Backend'], linestyle='--', color='blue', label='Mean Backend')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Performance')
plt.title('Run Chart for Manufacturing Sector')
plt.legend()

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
